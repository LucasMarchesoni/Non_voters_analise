#### **Assumindo que a amostra representa adequadamente a população americana, qual a raça que possui maior poder aquisitivo? E menor?**

Obs: Como não possuimos a renda, utilizaremos a mediana da faixa salarial

Separando num dataframe menor, agrupando pela raça e faixa salarial e filtrando para pegar as faixas de maiores e menores rendas.

```python
df_income_cat_hue_race = df[['income_cat', 'race']]
df_income_cat_hue_race = df.groupby(['race', 'income_cat']).size().reset_index(name='count')
df_income_cat_hue_race = df_income_cat_hue_race[(df_income_cat_hue_race['income_cat'] == "$125k or more") | (df_income_cat_hue_race['income_cat'] == "Less than $40k")]
df_income_cat_hue_race
```

Plotando o gráfico para enxergar a distribuição.

```python
_ = plt.figure(figsize=(10,7))
_ = plt.title("Distribuição por faixa salarial agrupada por raça")
_ = plt.xlabel("Faixa salarial")
_ = plt.ylabel("Total")
_ = sns.barplot(data=df_income_cat_hue_race, x="income_cat", y="count", hue="race")
```

Gerando a tabela com a frequência relativa.

```python
df_income_cat_hue_race['frequencia_relativa_pela_faixa_salarial (%)'] = (df_income_cat_hue_race['count'] / df_income_cat_hue_race.groupby('income_cat')['count'].transform('sum')).round(2) * 100
df_income_cat_hue_race['frequencia_relativa_pela_raça (%)'] = (df_income_cat_hue_race['count'] / df_income_cat_hue_race.groupby('race')['count'].transform('sum')).round(2) * 100

df_income_cat_hue_race
```

Na faixa salarial de 125 mil doláres ou mais, predomina a raça branca com 71% daqueles que responderam a pesquisa. E na faixa de menos de 40 mil dólares com 55% daqueles que responderam também predomina a raça branca. As raças hispanicas e negras, possuem mais pessoas que responderam na faixa daqueles que ganham menos de 40 mil dólares no ano.
Analisando apenas pela raça, a que possui mais pessoas na faixa salarial que recebem de 40 mil dólares ou menos com 72% é a raça negra, que também é que a que possui menos pessoas na faixa salarial de 125 mil dólares com apenas 25% das pessoas que responderam a pesquisa.

### 3.2 - O que é considerado ser um americano?

Glossário das respostas:

1.​ Muito importante

2.​ Pouco importante

3.​ Não tão importante

4.​ Muito pouco importante

#### **Um nível educacional maior implica em maior responsabilidade em relação a votar?**

Separando num dataframe menor, agrupando pela pergunta e nível educacional.

```python
df_vote_responsability = df[['Q2_1', 'educ']]
df_vote_responsability = df_vote_responsability.groupby(['Q2_1', 'educ']).size().reset_index(name='count')
df_vote_responsability
```

Plotando o gráfico para poder visualizar os dados.

```python
_ = plt.figure(figsize=(12,10))
_ = plt.title("Importancia voto eleições agrupada por nível educacional")
_ = plt.xlabel("Importância de votar nas eleições")
_ = plt.ylabel("Total")
_ = sns.barplot(data=df_vote_responsability, x="Q2_1", y="count", hue="educ")
```

Gerando a tabela com a frequência relativa pela pergunta para termos os dados quantitativos.

```python
df_vote_responsability['frequencia_relativa (%)'] = (df_vote_responsability['count'] / df_vote_responsability.groupby('Q2_1')['count'].transform('sum')).round(2) * 100

df_vote_responsability
```

Podemos afirmar que um nível maior de escolaridade implica numa importância maior de votar,  cerca de 43% daqueles que considerar muito importante votar são aqueles que possuem um diploma universitário, já aqueles que considerar pouco importante votar, cerca de 59% são aqueles que possuem apenas ensino médio ou menos.

#### **Dentre os não-brancos, o que é mais importante para ser um bom americano: votar em eleições, demonstrar a bandeira americana ou apoiar o exército? Existe uma diferença na preferência de não-brancos de maior poder aquisitivo vs menor poder aquisitivo?**

Separando em um dataframe menor com as informações necessárias

```python
df_race_non_white = df[['race', 'income_cat', 'Q2_1', 'Q2_4', 'Q2_7']]
df_race_non_white = df_race_non_white[df_race_non_white['race'] != 'White']
df_race_non_white
```

Plotando os gráficos lado a lado para poder identificar melhor as tendências.

Com relação ao apoio ao exército, todas as raças acreditam que sejam muito importante apoiar o exército.

```python
sns.set(style="whitegrid")

fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

for i, col in enumerate(['Q2_1', 'Q2_4', 'Q2_7']):
    sns.countplot(x='income_cat', hue=col, data=df_race_non_white, ax=axs[i], palette='husl')
    axs[i].set_title(f'Pergunta {col} com respostas agrupadas pela faixa salarial')
    
plt.tight_layout()

plt.show()
```

Com relação ao apoio ao exército, todas as raças não brancas independente da faixa salarial acreditam que seja muito importante apoiar o exército.

#### **Considerando que respeitar a opinião dos outros (Q2_8) é um indício de tolerância, ao compararmos as gerações de 18-30 anos, 31-50 anos e +51, qual é a mais tolerante e a menos tolerante?**

```python
df_per_generation = df[['ppage', 'Q2_8']]
df_per_generation
```

```python
def create_age_divisor(df, column):
  for i in range(len(column)):
    if column[i] >= 18 and column[i] <= 30:
      df.loc[i, 'age_divisor'] = "18-30"
    elif column[i] > 30 and column[i] <= 50:
      df.loc[i, 'age_divisor'] = "31-50"
    elif column[i] > 50:
      df.loc[i, 'age_divisor'] = "50+"

create_age_divisor(df_per_generation, df_per_generation['ppage'])

df_per_generation
```
